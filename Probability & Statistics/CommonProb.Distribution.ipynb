```python
%pip install scipy
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.linspace(-5, 5, 1000)
normal_pdf = stats.norm.pdf(x, loc=0, scale=1)
uniform_pdf = stats.uniform.pdf(x, loc=-2, scale=4)
poisson_pmf = stats.poisson.pmf(np.arange(0, 10), mu=3)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, normal_pdf, label="Normal")
plt.plot(x, uniform_pdf, label="Uniform")
plt.title("PDFs")
plt.legend()

plt.subplot(1, 3, 2)
plt.bar(np.arange(0, 10), poisson_pmf, color="orange")
plt.title("Poisson PMF")

plt.subplot(1, 3, 3)
plt.stem([0,1], [0.3, 0.7], use_line_collection=True)
plt.title("Bernoulli PMF (p=0.7)")

plt.tight_layout()
plt.show()
```
