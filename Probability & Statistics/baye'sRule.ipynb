```python
# Conditional Probability Example
def conditional_probability(P_A_and_B, P_B):
    return P_A_and_B / P_B

# Bayes' Theorem
def bayes_theorem(P_B_given_A, P_A, P_B):
    return (P_B_given_A * P_A) / P_B

# Example usage
P_A_and_B = 0.12
P_B = 0.3
print("P(A|B) =", conditional_probability(P_A_and_B, P_B))

P_B_given_A = 0.8
P_A = 0.3
P_B = 0.5
print("P(A|B) using Bayes’ Theorem =", bayes_theorem(P_B_given_A, P_A, P_B))
```
