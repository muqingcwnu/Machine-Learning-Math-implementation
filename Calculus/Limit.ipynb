```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)
y = np.where(x != 1, (x**2 - 1)/(x - 1), 0)  # Discontinuous at x = 1

plt.plot(x, y, label='f(x)')
plt.axvline(1, color='red', linestyle='--', label='x=1')
plt.legend()
plt.title('Visualizing Limit Behavior at x = 1')
plt.grid()
plt.show()
```
