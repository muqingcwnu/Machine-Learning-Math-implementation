```python
import sympy as sp

# Define symbols
x = sp.symbols('x')

# Example 1: Product Rule - f(x) = x^2 * sin(x)
f1 = x**2 * sp.sin(x)
f1_prime = sp.diff(f1, x)

# Example 2: Quotient Rule - f(x) = (x^2) / (x + 1)
f2 = x**2 / (x + 1)
f2_prime = sp.diff(f2, x)

# Example 3: Higher-Order Derivative - f(x) = x^3
f3 = x**3
f3_second_derivative = sp.diff(f3, x, 2)

print(f"Product Rule: f'(x) for x^2 * sin(x) = {f1_prime}")
print(f"Quotient Rule: f'(x) for (x^2) / (x+1) = {f2_prime}")
print(f"Second Derivative: f''(x) for x^3 = {f3_second_derivative}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic loss function: f(x) = x^2 - 4x + 3
def loss_function(x):
    return x**2 - 4*x + 3

# Derivative of the loss function: f'(x) = 2x - 4
def gradient(x):
    return 2*x - 4

# Gradient Descent implementation
def gradient_descent(learning_rate, num_iterations, initial_x):
    x = initial_x
    history = []
    
    for _ in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad  # Update rule
        history.append(x)
        
    return x, history

# Run Gradient Descent
final_x, history = gradient_descent(learning_rate=0.1, num_iterations=20, initial_x=0)

# Plot the loss function and the descent path
x_values = np.linspace(-2, 6, 100)
y_values = loss_function(x_values)

plt.plot(x_values, y_values, label="Loss Function (f(x) = x^2 - 4x + 3)")
plt.scatter(history, [loss_function(x) for x in history], color='red', label='Gradient Descent Steps')
plt.title("Gradient Descent on f(x) = x^2 - 4x + 3")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Minimum value of x found: {final_x}")
```

```python
import sympy as sp

# Define symbols for x and y
x, y = sp.symbols('x y')

# Define a function f(x, y) = x^2 + y^2
f = x**2 + y**2

# Partial derivatives
partial_x = sp.diff(f, x)
partial_y = sp.diff(f, y)

# Gradient of the function
gradient = [partial_x, partial_y]

print(f"Partial derivative with respect to x: {partial_x}")
print(f"Partial derivative with respect to y: {partial_y}")
print(f"Gradient: {gradient}")
```

```python
import sympy as sp

x, y = sp.symbols('x y')
f1 = x**2 + y
f2 = x * y
F = sp.Matrix([f1, f2])  # Vector-valued function
vars = sp.Matrix([x, y]) # Input vector

# Jacobian
J = F.jacobian(vars)
print("Jacobian Matrix:")
sp.pprint(J)
```

```python
import sympy as sp

x, y = sp.symbols('x y')
f = x**3 + y**2 + x*y

H = sp.hessian(f, (x, y))
print("Hessian Matrix:")
sp.pprint(H)
```

```python
import sympy as sp

x, y = sp.symbols('x y')
P = -y
Q = x
F = [P, Q]

div_F = sp.diff(P, x) + sp.diff(Q, y)
curl_F = sp.diff(Q, x) - sp.diff(P, y)

print("Divergence:", div_F)
print("Curl:", curl_F)
```

```python
import sympy as sp

# Define symbols
x, y, λ = sp.symbols('x y λ')

# Define the objective function and constraint
f = x**2 + y**2
g = x + y - 1  # Constraint: x + y = 1 ⟺ g(x, y) = 0

# Define the Lagrangian
L = f - λ * g

# Compute partial derivatives
grad_x = sp.diff(L, x)
grad_y = sp.diff(L, y)
grad_lambda = sp.diff(L, λ)

# Solve the system of equations
solutions = sp.solve([grad_x, grad_y, grad_lambda], (x, y, λ))

print("Solutions for Lagrange Multipliers:")
print(f"x: {solutions[x]}, y: {solutions[y]}, λ: {solutions[λ]}")


```
