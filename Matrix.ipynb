```python
import numpy as np
a=np.array([[1,2],[4,5]])
print(a)
print(a.shape)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = A + B
D = 2 * A
print("A + B:\n", C)
print("2 * A:\n", D)
```

```python
transposeA=A.T
print(transposeA)

transpose=np.transpose(A)
print(transpose)

def matrixTranspose(A):
    return np.transpose(A)
print(matrixTranspose(A))

```

```python
mul=np.dot(A,B)
print(mul)

mul2=A@B
print(mul2)

def matrixMultiplication(A, B):
    return np.dot(A, B)
print(matrixMultiplication(A, B))
```

```python
IdM=np.eye(2)
print(IdM)

IDM=np.identity(2)
print(IDM)

IDM2=np.eye(3,3)
print(IDM2)

def identityMatrix(n):
    return np.eye(n)
print(identityMatrix(2))
```

```python
invM=np.linalg.inv(A)
print(invM)

def inverseMatrix(A):
    return np.linalg.inv(A)
print(inverseMatrix(A))
```

```python
squareMat=np.array([[1, 2], [3, 4]])

trace=np.trace(squareMat)
print(trace)

def matrixTrace(matrix):
    return np.trace(matrix)
print(matrixTrace(squareMat))

```

```python
det=np.linalg.det(squareMat)
print(det)

def matrixDeterminant(matrix):
    return np.linalg.det(matrix)
print(matrixDeterminant(squareMat))


```

```python
eigenvalues, eigenvectors = np.linalg.eig(squareMat)
print(eigenvalues)
```
