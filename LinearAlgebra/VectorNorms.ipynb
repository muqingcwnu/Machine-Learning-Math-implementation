```python
import numpy as np 
v1=np.array([1,2,4])
l1Norm=np.linalg.norm(v1,ord=1)
print(l1Norm)


def l1Norm(v1):
    return np.sum(np.abs(v1))

print(l1Norm(v1))
```

```python

L2form=np.linalg.norm(v1,ord=2)
print(L2form)
def l2Norm(v1):
    return np.sqrt(np.sum(v1**2))
print(l2Norm(v1))
```

```python
linfNorm=np.linalg.norm(v1,ord=np.inf)
print(linfNorm)
def linfNorm(v1):
    return np.max(np.abs(v1))
print(linfNorm(v1))
```

```python
lpNorm=np.linalg.norm(v1,ord=3)
print(lpNorm)
def lpNorm(v1): 
    return np.sum(np.abs(v1)**3)**(1/3)
print(lpNorm(v1))
```

```python

frobeniousNorm=np.linalg.norm(v1, ord=2)
print(frobeniousNorm)
def frobeniousNorm(v1):
    return np.sqrt(np.sum(v1**2))
print(frobeniousNorm(v1))
```

```python
v= np.array([1, 2, 3])

unitVector = v / np.linalg.norm(v)
print(unitVector)

def unitVector(v):
    norm = np.linalg.norm(v)
    if norm == 0:       
        return v
    return v / norm

unit = unitVector(v)
print(unit)
```
