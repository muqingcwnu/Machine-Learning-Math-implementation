```python
import numpy as np
import matplotlib.pyplot as plt 

A=np.array([[1, 2], [3, 4]])
B=np.array([[5, 6], [7, 8]])
AB=A@B
print(AB)

plt.quiver(A[0, 0], A[1, 0], B[0, 0], B[1, 0], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(A[0, 1], A[1, 1], B[0, 1], B[1, 1], angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.legend()
plt.title('Vector Representation of Matrices A and B')
plt.gca().set_aspect('equal')
plt.show()
```

```python
v = np.array([3, 4])
u = np.array([1, 2])
proj_v_on_u=(np.dot(v,u)/np.dot(u,u))*u
print(u)
print(v)
print(proj_v_on_u)
```

```python
plt.quiver(0, 0, v[0], v[1], color='r', angles='xy', scale_units='xy', scale=1, label='v')
plt.quiver(0, 0, u[0], u[1], color='b', angles='xy', scale_units='xy', scale=1, label='u')
plt.quiver(0, 0, proj_v_on_u[0], proj_v_on_u[1], color='g', angles='xy', scale_units='xy', scale=1, label='proj_v_on_u')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.axhline(0, color='black', linewidth=0.3)
plt.axvline(0, color='black', linewidth=0.3)
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Projection of v onto u")
plt.show()
```
